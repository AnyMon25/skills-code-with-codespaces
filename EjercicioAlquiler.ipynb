Formas de selección especifica o particiones

Excluyendo valores nulos

```python
datos = pd.read_csv("alquiler_residencial.csv", sep=";")
datos.head(10)
```

Ambos metodos identifican los valores nulos dentro del dataframe isnull transforma los valores NaN en true y la función notnull devuelve como false los valores NaN.

```python
datos.isnull()
```

```python
datos.notnull()
```

La función .info devuelve el previuw general de los datos, la previsualización ayuda a identificar las columnas del dataframe donde existen datos nulos

```python
datos.info()
```

La traducción seria, dentro del dataframe general o total, busca dentro de Datos la columna Valor y devuelve las filas deonde exista valores NaN dentro de la columna Valor

```python
datos[datos["Valor"].isnull()]
```

Dropna elimina los valores NaN o Nulos del dataframe pero solo durante la consulta, para hacer el remove valido dentro de la dataframe original o raiz se utiliza la funcion inplace = True para hcer los cambios directamente en el dataframe

```python
A = datos.shape[0]
datos.dropna(subset=["Valor"], inplace= True)
B = datos.shape[0]
A - B
```

Creando criterios de selección, en este ya involucra enunicados mas complejos de mas de una variable a tomar en cuenta, en este ejemplo Mantenimiento y Tipo de vivienda

```python
datos[datos["Mantenimiento"].isnull()].shape[0]
```

Selecciona las filas que contengan doble valor NaN nulos, tipo de vivienda = Nulo
mantenimiento = Nulo
Doble selección de criterios

```python
seleccion = (datos["Mantenimiento"].isnull()) & (datos["Tipo"]=="Departamento")
```

```python
A = datos.shape[0]
datos= datos[~seleccion]
B = datos.shape[0]
A - B
```

```python
datos[datos["Mantenimiento"].isnull()].shape[0]
```

Con la función fillna, busca todos los valores nulos y los sustituye por un valor determinado

En este caso se llevo acabo en dos columnas que presentaban números nulos, la columna namtenimiento e impuestos, es un diccionario porque incorpora la estructura Llave y Valor

```python
datos = datos.fillna({"Mantenimiento":0, "Impuesto":0})
```

```python
datos.info()
```

```python
datos.to_csv("alquiler_residencial.csv", sep=";", index= False)
```

Creando nueva variable, el Valor Bruto siendo la suma del valor, mas el matenimiento y el impuesto

```python
datos["Valor Bruto"] = (datos["Valor"]) + (datos["Mantenimiento"]) + (datos["Impuesto"])
datos.head(10)
```

Mismo proceso de creación de variable, lavor del metro cuadrado equivale al valor entre el area en metros cuadrados

```python
datos["Valor metro cuadrado"] = (datos["Valor"]) / (datos["Area"])
datos.head(10)
```

round = redondear cifras

```python
datos["Valor metro cuadrado"] = datos["Valor metro cuadrado"].round(2)
datos.head(10)
```

```python
datos["Valor bruto m2"] = (datos["Valor Bruto"]) / (datos["Area"])
datos.head(10)
```

```python
datos["Valor bruto m2"] = datos["Valor bruto m2"].round(2)
datos
```

Creando agrupamientos

```python
datos = pd.read_csv("alquiler_residencial.csv", sep=";")
datos.head(10)
```

```python
barrios = ["Ate", "San Borja", "Comas"]
seleccion = datos["Distrito"].isin(barrios)
datos = datos[seleccion]
```

```python
barrios
```

```python
datos.head(10)
```

```python
datos["Distrito"].drop_duplicates()
```

```python
grupo_barrios = datos.groupby("Distrito")
```

```python
type(grupo_barrios)
```

```python
for barrios, data in grupo_barrios:
  print(data)
```

```python
grupo_barrios[["Valor", "Mantenimiento"]].mean().round(2)
```

Estadistica descriptiva

Función describe, describe el conjunto de datos y arroja datos estadisticos relecantes como la media, total y cuartiles

```python
grupo_barrios["Valor"].describe().round(2)
```

```python
grupo_barrios["Valor"].aggregate(["min", "max"]).round(2).rename(columns={"min":"minimo","max":"maximo"})
```

Importando la bilbioteca matplotl que sirve para graficar los datos procesados por pandas

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc("figure", figsize = (20,10))
```

```python
fig = grupo_barrios["Valor"].mean().plot.bar(color= "blue")
fig.set_ylabel("Valor del alquiler")
fig.set_title("Valor medio del alquiler por distrito", {"fontsize":22})
```
